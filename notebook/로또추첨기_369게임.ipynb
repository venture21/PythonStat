```python
#3,6,9게임
#1~100까지 숫자를 출력하는데 3,6,9가 들어간 숫자가 나오면 출력 '박수'

# 입력받은 문자열을 정수로 변환해서 n에 저장
n = int(input("숫자를 입력하시오:"))

for i in range(1,n+1):
    #숫자를 문자열로 변환
    myStr = str(i)

    # 3,6,9를 카운트하는 변수
    count = 0
    # 문자열 '12' -> '1', '2'
    for s in myStr:
        if s=='3' or s=='6' or s=='9':
            count+=1
            print('박수',end=' ')
            
    if count>0:
        # for문을 빠져나오면 줄바꿈을 하는 것으로
        print("")
    else:
        print(myStr)
```
